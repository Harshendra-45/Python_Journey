```python
#64. Count Frequencies of vowel
str1 = input("Enter a string:")
a,e,i,o,u=0,0,0,0,0
print(str1)
for j in str1:
    if j=="a" or j=="A":
        a+=1
    elif j=="e" or j=="E":
        e+=1
    elif j=="i" or j=="I":
        i+=1
    elif j=="o" or j=="O":
        o+=1
    elif j=="u" or j=="U":
        u+=1
print("a",a,"e",e,"i",i,"o",o,"u",u)

```
