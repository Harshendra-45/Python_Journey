```python
# 7 convert to lowercase
str = input("Enter a string: ")
pas = ""
for i in str:
    if i.islower():
        pas+=i
    else:
        pas+=i.lower()
print(pas)
```

```python
#10. Trim leading, trailing, or extra spaces from a string.
str=input("Enter a string: ")
res=""
for i in range(len(str)-1):
    if str[i]=="." and str[i+1]==".":
        pass
    elif str[i]=="." and str[i-1]!=".":
        pass
    else:
        res+=str[i]
print(res)
```
