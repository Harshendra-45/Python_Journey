```python
class Employee:
    def __init__(self,id,name,sal):
        self.id = id
        self.name = name
        self.sal = sal
    def calculations(self):
        self.hra = 0.20*self.sal
        self.da = 0.15*self.sal
        self.gross = self.sal+self.hra+self.da
    def display(self):
        print("------- Employee Salary Details ------")
        print("Employee ID      : ",self.id)
        print("Employee Name    : ",self.name)
        print("Basic Salary     : ",self.sal)
        print("HRA              : ",self.hra)
        print("DA               : ",self.da)
        print("Gross Salary     : ",self.gross)
e1 = Employee("E101","Rahul",50000.0)
e1.calculations()
e1.display()
```

```python
class Customer:
    def __init__(self,id,name,units):
        self.customer_id=id
        self.customer_name=name
        self.units_consumed=units
    def calculations(self):
        self.total_bill = (self.units_consumed*8)+150
    def display(self):
        print("------ Electricity Bill ------")
        print("Customer ID       : ",self.customer_id)
        print("Customer Name     : ",self.customer_name)
        print("Units Consumed    : ",self.units_consumed)
        print("Total Bill Amount : ",self.total_bill)
c1 = Customer("C101","Amit Verma",350.0)
c1.calculations()
c1.display()
```

```python
class Product:
    def __init__(self,id,name,qn,ppi):
        self.product_id=id
        self.product_name=name
        self.product_quantity=qn
        self.product_price_peritem=ppi
    def bill(self):
        self.total_bill=self.product_quantity*self.product_price_peritem
        if self.total_bill>5000:
            self.discount = 0.10*self.total_bill
            self.final_bill=self.total_bill-self.discount
        else:
            self.discount = 0.05*self.total_bill
            self.final_bill=self.total_bill-self.discount
    def display(self):
        print("------ Shopping Bill ------")
        print("Product ID        : ",self.product_id)
        print("Product Name      : ",self.product_name)
        print("Quantity          : ",self.product_quantity)
        print("Price Per Item    : ",self.product_price_peritem)
        print("Total Amount      : ",self.total_bill)
        print("Discount          : ",self.discount)
        print("Final Amount      : ",self.final_bill)
pr1=Product("P101","Laptop",2,35000)
pr1.bill()
pr1.display()
```

```python
class Student:
    def __init__(self,roll_no,stu_name,m1,m2,m3):
        self.roll_no = roll_no
        self.student_name = stu_name
        self.marks1 = m1
        self.marks2 = m2
        self.marks3 = m3
    def calc(self):
        self.total_marks =self.marks1+self.marks2+self.marks3
        self.percentage = self.total_marks/3
    def grade(self):
        if self.percentage>=90:
            self.grade="A"
        elif 75<=self.percentage<=89:
            self.grade="B"
        elif 60<=self.percentage<=74:
            self.grade="C"
        else:
            self.grade="D"
    def display(self):
        print(f"------ Student Result ------\n\
        Roll Number      : {self.roll_no}\n\
        Student Name     : {self.student_name}\n\
        Total Marks      : {self.total_marks}\n\
        Percentage       : {self.percentage}\n\
        Grade            : {self.grade}")
        
roll_no = int(input("Enter roll no: "))
name = (input("Enter name: "))
mark1= int(input("Enter marks1 : "))
mark2= int(input("Enter marks2 : "))
mark3= int(input("Enter marks3 : "))

d=Student(roll_no,name,mark1,mark2,mark3)
d.calc()
d.grade()
d.display()
```

```python
class Guest:
    def __init__(self,id,name,days,rent):
        self.id=id
        self.name=name
        self.days=days
        self.rent=rent
    def bill(self):
        self.room_bill = self.days*self.rent
        self.gst=0.12*self.room_bill
        self.total_bill=self.room_bill+self.gst
    def display(self):
        print("------ Hotel Bill ------")
        print("Guest ID              : ",self.id)
        print("Guest Name            : ",self.name)
        print("No of days            : ",self.days)
        print("Room charge per day   : $",self.rent)
        print("Room bill             : $",self.room_bill)
        print("GST (12%)             : $",self.gst)
        print("Final Bill            : $",self.total_bill)
id = input("enter guest id:")
name = input("enter guest name:")
days=int(input("Enter no of days: "))
rent=int(input("Enter charge for 1 day"))
g1=Guest(id,name,days,rent)
g1.bill()
g1.display()
```

```python
class Book:
    def __init__(self,book_id,title,aut):
        self.book_id = book_id
        self.title=title
        self.author=aut
        self.status="Available"
    def display_details(self):
        print("\n------ Book Details ------")
        print("Book ID      : ",self.book_id)
        print("Title        : ",self.title)
        print("Author       : ",self.author)
        print("Status       : ",self.status)
    def issue_book(self):
        self.status="Issued"
        print("\nBook Issued Successfully")
        print("------ Book Details ------")
        print("Book ID      : ",self.book_id)
        print("Title        : ",self.title)
        print("Author       : ",self.author)
        print("Status       : ",self.status)
        
    def return_book(self):
        self.status="Available"
        print("\nBook Returned Successfully")
        print("------ Book Details ------")
        print("Book ID      : ",self.book_id)
        print("Title        : ",self.title)
        print("Author       : ",self.author)
        print("Status       : ",self.status)

id = input("Enter book id : ")
title = input("Enter book title : ")
author = input("Enter author name : ")
b1 = Book(id,title,author)
b1.issue_book()
b1.display_details()
b1.return_book()
```
