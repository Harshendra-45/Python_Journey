### Print name n times .


```python
c = 0
def show(name,n):
        global c
        if  c == n:
            return 
        print(name) 
        c+=1
        show("Harsh",n)
show("Harsh",5)
```

### Print linearly from 1 to n



```python
c = 1
def show(n):
    global c 
    if c==n+1:
        return 
    print(c)
    c+=1
    show(n)
n = int(input("Enter a num"))
show(n)
```

### Print form n to 1


```python
def show(n):
    if n==0:
        return 
    print(n)
    n-=1
    show(n)
n=int(input("Enter a no: "))
show(n)
```

### Sum of n natural no


```python
def show(n):
    if n==0:
        return 0
    return n + show(n-1)
n = int(input("Enter a no: "))
show(n)


```
