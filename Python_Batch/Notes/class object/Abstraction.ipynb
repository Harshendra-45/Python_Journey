# Abstraction:
It is an OOP concept that means hiding internal implementation details and showing only essential features of an object to the user. 
Python provides abstraction using abstract classes and methods. For this purpose python uses abstract method modules and abc module. 

### Abstract class:


It is a class that can not be instantiated, means its object cannot be created, It acts as a blueprint to other classes. 

```python
from abc import ABC,abstractmethod
class Bank(ABC):
    @abstractmethod
    def intrest(self):
        pass 
obj=Bank()
obj.interest() # Error as it cant be instantiated
```

#### abc :
It stand for Abstract base class, its a builtin module in python that helps us to create abstract class and  abstract methods. 
##### ABC:
it is a base class used to create abstract class. 
#### Abstract method:
It is a decorator used to declare a method without implementation, it force the child class to implement that method. 

```python
from abc import ABC,abstractmethod
class Bank(ABC):
    @abstractmethod
    def interest(self):
        pass 
class SBI(Bank):
    def interest(self):
        print("Sbi interest, 8.0 ")
class HDFC(Bank):
    pass

obj=SBI()
obj.interest()
# obj1=HDFC() # It is compulsory to add interest as a method here 
# obj1.interest()
```

```python
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*self.r*self.r
class rectangle(Shape):
    pass #here area is mandatory otherwise its object will not be created

obj=Circle(5)
print(obj.area())
```

```python
from abc import ABC,abstractmethod
class Employee(ABC):
    def __init__(self,name,salary):
        self.name=name
        self.salary = salary
    @abstractmethod
    def calculatebonus(self):
        pass
    def display(self):
        print("Employee name", self.name)
        print("salary", self.salary)
class Manager(Employee):
    def calculatebonus(self):
        bonus=self.salary*0.20
        print("manager bonus",bonus)
class Developer(Employee):
    def calculatebonus(self):
        bonus=self.salary*0.10
        print("Developer bonus",bonus)
obj1=Manager("Anil",5000000)
obj1.calculatebonus()
obj1.display()
obj2=Developer("Sahil",50000)
obj2.calculatebonus()
obj2.display()
```

#### Advantages of Abstract class
1. Enforce methof implementation: Abstract classes force child class to implement certain method. This ensures all subclasses follow same structure.
2. Reduce Complexity: User only interacts with essesntial features without worrying about internal implementation 
3. Supports Loose Coupling: 
4. It provides common interface
