### Method Overriding
It means a child class provides its own implementation of a method that is already defined in the parent class, In this concept the method name in the child class must be same as of parent class

```python
class Parent:
    def calculation(self):
        print("Parent")
class Child(Parent):
    def calculation(self):
        print("Child")
obj1=Child()
obj1.calculation()
```

#### Super in overriding


```python
class Parent:
    def calculation(self):
        print("Parent")
class Child(Parent):
    def calculation(self):
        super().calculation()
        print("Child")
obj1=Child()
obj1.calculation()
```

In case of overriding :
* Method name  and parameter should be same in parent and child class 
* Inheritance is Required


```python
class Employee:
    def salary(self,basic,bonus):
        print("total ", bonus+basic)
class Manager(Employee):
    def salary(self, basic, bonus):
       print("Manager salary", basic+bonus+5000)
obj1=Manager()
obj1.salary(50000,3000)
```

### Dynamic Polymorphism(Runtime):
It is a type of polymorphism where the method to be executed is determined at runtime, rather than at compile time. 
In Python Dynamic polymorphism is achieved through 
1. Method overriding
2. Duck Typing: Its a concept in python where the type is object is not checked instead python checks whether the object has required method or behaviour. 


```python
# Duck Typing
class Bird:
    def fly(self):
        print("Bird can fly")
class Airplane:
    def fly(self):
        print("Airplane can fly")
objects=[Bird(),Airplane()]
for obj in objects:
    obj.fly()
```

### Builtin Polymorphism:
It means that python predefine functions can work with different types of objects and behave differently depending on the object type 


```python
print(len("Abcddddddd"))
print(len([1,1,1,1,1,1,1,]))
```

#### Operator Polymorphism:
It means the same operator behave differently depending on the operand.

```python
print(10+20)
print("wel"+"come")
print(5*3)
print("wel"*3)
```
