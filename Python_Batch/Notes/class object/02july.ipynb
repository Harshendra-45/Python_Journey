#Self
In python self is a fundamental concept when working with oops. It represents instance of a class being used. 
Whenever we create object form class sels refers to the current object. When we call a method or an object then self automatically gets passed to the method. 
self allows us to modify the internal state of an object


````python
#self ex1
class Counter:
    def __init__(self):
        self.count = 0
    def increment(self):
        self.count+=1
    def decrement(self):
        self.count-=1
    def getcount(self):
        return self.count
c = Counter()
c.increment()
c.increment()
print(c.getcount())
````

## Types of variables in a class
In Python OOPs , variables inside a class are mainly classified based on where they are created and how they are shared.
There are many types:
1. Instance variable
2. Class Variable 
3. Local Variable
### Instance Variable
These are variable who have separate copy , created for every object. They belong to individual object not the class.

````python
#Instance ex:
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

s1 =Student("Deepika",30)
s2 =Student("Radha",22)
print(s1.name)
print(s2.name)
s1.name="diya"
print(s1.name)
````

### Instance characteristics
* we can access instance variable within the class by using self, and outside the class by using reference variable .
* Their values differ from object to object. 
* they are stored in object memory


we can create Instance variable inside constuctor or inside functions.


````python
#inside function instance
class Student:
    def set(self,name,age):
        self.name=name
        self.age = age
s1=Student()
s2=Student()
s1.set("Diya",21)
print(s1.name)
````

we can create instance variable outside the class also


````python
class Student:
    pass

s1=Student()
s1.name="Diya"
print(s1.name)
````

### Class Variables
These are variables shared by all object of a class.  
Only one copy exist for entire class.  
They are declared inside the class but outside the method, they are accessed using class name

````python
#Class variable example
class Student:
    college="IIT Delhi"
    def __init__(self,name):
        self.name=name

s1=Student("Deepika")
s2=Student("Virat")

print(s1.name, s1.college)
print(s2.name, s2.college)
#After changing class variable college value
Student.college="Sait"
print(s1.name, s1.college)
print(s2.name, s2.college)
````

### Types of Methods in Python
Python provides 3 types of methods:
1. Instance method
2. Class method
3. Static method
#### 1. Instance Method  
It works with object. It can modify or access object data . It uses self parameter , its called using object. Examples we have done above.
##### Types of Instance Method:
1. Accessor method(getter): It is used to read data.  
    ```python
    def get_marks(self):  
            return self.marks
    ```
2. Mutator method(setter):used to modify data. 
   ```python 
    def set_marks(self):  
            self.marks=marks
    ```
#### 2. Class Method
It works with class variables instead of instance variables. It affects the whole class, it is useful when we have to change class variable. It uses cls parameter, they are decorated with @classmethod.They are called using class name.
```python
class Student:
    college="NIT"
    def __init__(self,name):
        self.name=name

    @classmethod
    def change_college(cls,new):
        cls.college=new
s1=Student("deepika")
print(s1.name)
print(s1.college)
Student.change_college("Sait")
print(s1.college)
```
Cls:  
Cls represents the class itself, it refers to current class. It gets automatically passed as a first parameter to a class method any other name is also possible.  
@classmethod is a decorator used to define class method. It belongs to class not object. It can access class variables , it can modify class state.
#### 3. Static method
Static method belong to class and doesn't access class data. It works like a normal function but placed inside a class for organisation.  
Here **staticmethod** decorator is used. A static method is a utility function inside a class that doesn't depend on object or class data.  
We can call static method either by using class name or by using object name.


````python
#Class method example
class Student:
    college="NIT"
    def __init__(self,name):
        self.name=name

    @classmethod
    def change_college(cls,new):
        cls.college=new
s1=Student("deepika")
print(s1.name)
print(s1.college)
Student.change_college("Sait")
print(s1.college)
````

````python
#Static method example
class Calculator:
    @staticmethod
    def add(a,b):
        return a+b
c1=Calculator()
print(c1.add(10,20))
print(Calculator.add(10,20))
print(c1.__dict__)
print(Calculator.__dict__)
````

#### Method Chaining
It's a concept when one method of a class calls another method of the same class. There are 2 patterns for this  
1. Internal Method Call: calling another instance method internally using self.
2. Method chaining using returning self: In example every method return self which allows next method to be called in the same line , with the help of this we can call multiple methods in a single line 



````python
#Internal method call 
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def displayname(self):
        print("Name is", self.name)
    def displaymarks(self):
        print("Marks",self.marks)
    def displayall(self):
        self.displayname()
        self.displaymarks()
s1 = Student("deepika",90)
s1.displayall()
````

````python
#Returning self
class Student:
    def __init__(self,name):
        self.name=name
        self.marks=0
    def set_marks(self,marks):
        self.marks=marks
        return self
    def hello(self):
        print(f'Heyy {self.name}')
        return self
    def display(self):
        print(f'Your marks are : {self.marks}')
        #return self ## if we have to chain more methods
s1 = Student("deepika")
s1.hello().set_marks(80).display()
````

# Encapsulation 
Encapsulation is a binding of data with instance members .  
Encapsulation means wrapping data and functions inside a single unit called class.  
It is the process of binding data members and member functions together inside a class and control there access using access specifiers.  
With the help of encapsulation we can achieve data security.
### Access Specifiers: 
These define how class members can be accessed from outside the class.  
Types:
#### 1. Public  
Public members are accessed from anywhere , by default all members in python are public.
#### 2. Protected
Protected members are variable or methods, which are accessed within the class or subclass. In Python protected members are defined using **single underscore**(self._name)
#### 3. Private
Private Members can not be accessed directly from outside the class. They are used to restrict the data. Defined using **double underscore** (self.__salary)


````python
# Protected
class Student:
    def __init__(self,name):
        self._name=name
s1=Student("deepika")
print(s1._name) # we can access using _ . As python doesn't have any restriction on protected members
````

````python
# Private
class Student:
    def __init__(self,name,salary):
        self.name=name
        self.__salary=salary
s1=Student("deepika",500000)
print(s1.name)
#print(s1.__salary) ## Give error as salary is private
````

````python
# Setter & Getter method
class Student:
    def __init__(self,name,salary):
        self.name=name
        self.__salary=salary
    def get_salary(self):
        return self.__salary
    def set_salary(self,salary):
        self.__salary=salary
s1=Student("Deepika",50000)
print(s1.name)
print(s1.get_salary())
s1.set_salary(10)
print(s1.get_salary())
````

````python
class Bank_Balance:
    def __init__(self,bal):
        self.__bal = bal
    def deposit(self,bal):
        print("Deposit Successful")
        self.__bal = self.__bal+bal
    def withdraw(self,amt):
        if self.__bal>amt:
            print("withdraw successfull")
            self.__bal=self.__bal-amt
        else:
            print("Not enough Balance")
    def get_bal(self):
        return self.__bal
b1 = Bank_Balance(50)
b1.deposit(500)
print(b1.get_bal())
b1.withdraw(5000)
b1.withdraw(400)
print(b1.get_bal())
````
