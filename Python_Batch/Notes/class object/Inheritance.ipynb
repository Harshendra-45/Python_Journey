## Inheritance
 It is a fundamental concept in OOP that alllows a class  to acquire the properties and behaviours of another class called the parent class(base class).  
#### Advantages:
1. Code reusability
2. Improve maintainability
#### Types of Inheritance :
1. Single Inheritance: In this a child class inherit from the single parent class.


```python
#Single inheritance
class Parent:
    def fun1(self):
        print("This is parent class function")
class Child(Parent):
    def fun2(self):
        print("This is child class")
obj = Child()
obj.fun1()
obj.fun2()
```

```python
#Single
class Person:
    def __init__(self,name):
        self.name=name
class Employee(Person):
    def show_role(self):
        print(self.name,"is a employee")
obj=Employee("Deepika")
obj.show_role()
```

#### 2. Multilevel Inheritance: When a class is inherited from another child class, then it is called multilevel inheritance.  
    Base class  
    Intermediate class  
    Child class 


```python
class GrandParent:
    def fun1(self):
        print("Fun1 from grandparent")
class Parent(GrandParent):
    def fun2(self):
        print("Fun2 from Parent")
class Child(Parent):
    def fun3(self):
     print("Fun3 from child")
obj=Child()
obj.fun1()
obj.fun2()
obj.fun3()
```

Super  
    It is a built in function in python used in Inheritance, to access the methods and properties of the parent class from the child class.
    It helps a child class to reuse the functionality of its parent class.  
    Syntax:  
        super().methodname(arguments)  
        super().\_\_init\_\_(arguments)
        

```python
class Person:
    def __init__(self):
        print("person constructor is called")
class Employee(Person):
    def __init__(self):
        print("Employee constructor")
emp = Employee() ### Only child  constructor will be called 
```

```python
class Person:
    def __init__(self):
        print("person constructor is called")
class Employee(Person):
    def __init__(self):
        super().__init__()
        print("Employee constructor")
emp = Employee() # Both will get called as super is used 
```

```python
class Person:
    def __init__(self,name,age,address):
        self.name=name
        self.age=age
        self.address=address
        print("Parent class")
class Employee(Person):
    def __init__(self,name,age,address,salary):
        super().__init__(name,age,address)
        self.salary=salary
        print("child class")
emp=Employee("Deepika",30,"chennai",900000)
print(emp.name)
print(emp.age)
print(emp.address)
print(emp.salary)
```

```python
### When parent class and child class have a function of same name, then super again shows its magic and calls the function of parent class.
class Person:
    def show(self):
        print("Person method")
class Child(Person):
    def show(self):
        super().show()
        print("Child method")
c=Child()
c.show()
```

#### 3. Multiple Inheritance:  
If a child class inherits from more than one parent class, then it is called Multiple Inheritance.  

```python
class Father:
    def house(self):
        print("father has house")
class Mother:
    def laptop(self):
        print("mother has  laptop")
class Child(Father,Mother):
    def nothing(self):
        print("child has nothing")
c=Child()
c.house()
c.laptop()
c.nothing()
```

```python
class Father:
    def house(self):
        print("father has house")
    def laptop(self):
        print("Father has laptop")
class Mother:
    def laptop(self):
        print("mother has  laptop")
class Child(Father,Mother): ## imp
    def nothing(self):
        print("child has nothing")
c=Child()
c.house()
c.laptop() ## This will call father's laptop function as it is written first in  **imp**
c.nothing()
```

Constructors in Multiple inheritance : 

```python
class Father:
    def __init__(self):
        print("Father constructor")
class Mother:
    def __init__(self):
        print("Mothers constructor")
class Child(Father,Mother): ## imp
    def __init__(self):
        super().__init__() ## This will also follow name order so fathers constructor will be called
        print("child ")
c=Child()
```

#### **Mro order(Method Resolution Order)**
When multiple parent classes contains same method name or constructor python must decide which method to call first ,for this purpose python uses method resolution order  




#### Direct parent class call:  
 In this approach the child class calls the parent class method using the parent class name . 

```python
class Person:
    def __init__(self,name):
        self.name=name
        print("person constructor")
class Employee:
    def __init__(self,name,salary):
        Person.__init__(self,name)
        self.salary=salary
        print("Emp constructor")
obj=Employee("Deepika",90000)
```

Problems with Direct parent call:
1. Tight coupling: Here the child class is directly dependent on the parent class name. If the parent class name changes then code must be updated 
2. Doesnt work well with multiple inheritance bcoz direct method call ignores Mro.


```python
## 2nd Drawback of direct parent call example
class A:
    def show(self):
        print("A")
class B:
    def show(self):
        print("B")
class C(A,B):
    def show(self):
        A.show(self)
        B.show(self)
        print("C")
obj=C()
obj.show()
```

#### 4. Heirarchical Inheritance
In this type Multiple child classes inherit from the same parent class 

```python
class A:
    def show(self):
        print("A")
class B(A):
    def show2(self):
        print("B")
class C(A):
    def show3(self):
        print("C")
obj1=C()
obj1.show3()
obj1.show()
#obj1.show2() # This gives error as B and A have no relation
obj2 = B()
obj2.show2()
obj2.show()
```

#### 5. Hybrid Inheritance
It is a combination of 2 or more types of Inheritance. 

```python
class A:
    def show1(self):
        print("A")
class B(A):
    def show2(self):
        print("B")
class C(A):
    def show3(self):
        print("C")
class D(B,C):
    def show4(self):
        print("D")
obj1=D()
obj1.show1()
obj1.show2()
obj1.show3()
obj1.show4()
print(D.mro())
```
