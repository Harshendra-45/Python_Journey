#self
In python self is a fundamental concept when working with oops. It represents instance of a class being used. 
Whenever we create object form class sels refers to the current object. When we call a method or an object then self automatically gets passed to the method. 
self allows us to modify the internal state of an object


```python
#self ex1
class Counter:
    def __init__(self):
        self.count = 0
    def increment(self):
        self.count+=1
    def decrement(self):
        self.count-=1
    def getcount(self):
        return self.count
c = Counter()
c.increment()
c.increment()
print(c.getcount())
```

## Types of variables in a class
In Python OOPs , variables inside a class are mainly classified based on where they are created and how they are shared.
There are many types:
1. Instance variable
2. Class Variable 
3. Local Variable
### Instance Variable
These are variable who have separate copy , created for every object. They belong to individual object not the class.

```python
#Instance ex:
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age

s1 =Student("Deepika",30)
s2 =Student("Radha",22)
print(s1.name)
print(s2.name)
s1.name="diya"
print(s1.name)
```

### Instance characteristics
* we can access instance variable within the class by using self, and outside the class by using reference variable .
* Their values differ from object to object. 
* they are stored in object memory


we can create Instance variable inside constuctor or inside functions.


```python
#inside function instance
class Student:
    def set(self,name,age):
        self.name=name
        self.age = age
s1=Student()
s2=Student()
s1.set("Diya",21)
print(s1.name)
```

we can create instance variable outside the class also


```python
class Student:
    pass

s1=Student()
s1.name="Diya"
print(s1.name)
```

### Class Variables
These are variables shared by all object of a class.  
Only one copy exist for entire class.  
They are declared inside the class but outside the method, they are accessed using class name

```python
#Class variable example
class Student:
    college="IIT Delhi"
    def __init__(self,name):
        self.name=name

s1=Student("Deepika")
s2=Student("Virat")

print(s1.name, s1.college)
print(s2.name, s2.college)
#After changing class variable college value
Student.college="Sait"
print(s1.name, s1.college)
print(s2.name, s2.college)
```

####  \_\_dict\_\_ 
it represents instance variable of an object as dictionary(key and value pairs)

```python
class Student:
    college="IIT Delhi"
    def __init__(self,name,city):
        self.name=name
        self.city=city
        
s1=Student("Deepika","chennai")
s2=Student("Virat","Delhi")

print(s1.__dict__)
print(s2.__dict__)
#changing college of s1 object, here when changing objects class variable it creates a instance copy variable
s1.college="sait"

print(s1.__dict__)
print(s2.__dict__)
```

```python
class Test:
    x = 10
t1=Test()
t2=Test()
t1.x=90
print(t1.x)
print(t2.x)
print(t1.__dict__)
Test.x=100
print(t1.x)
print(t2.x)
Test.y=900
print(t1.y)
print(t2.y)
```
